```python
num_customers = 5

# Random seed for reproducibility
np.random.seed(0)

# Create a list to store data for each customer
data = []

# Generate data for each customer
for customer_id in range(1, num_customers + 1):
    # Randomly choose the number of orders for each customer (between 1 and 5)
    num_orders = np.random.randint(1, 6)
    
    # Generate data for each order
    for order_id in range(1, num_orders + 1):
        # Randomly choose the number of products in each order (between 1 and 10)
        num_products = np.random.randint(1, 11)
        
        # Append data for each product in the order to the list
        for product_id in range(1, num_products + 1):
            data.append([customer_id, order_id, product_id])

# Create a DataFrame from the list of data
df = pd.DataFrame(data, columns=['Customer_ID', 'Order_ID', 'Product_ID'])

print(df)
```
